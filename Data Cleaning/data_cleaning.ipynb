```python
data4 = pd.read_csv("data4.csv")
```

```python
data4.drop(["Words" , "S.No"] , axis = 1 , inplace = True)
```

```python
sentiments = [mode(eval(x).values()) for x in data4["Decisions"]]
```

```python
data4["Sentiments"] = sentiments
data4.drop(["Decisions"] , axis  = 1 , inplace = True)
```

```python
data4.rename(columns = {"Title" : "Sentence" , "Sentiments" : "Sentiment"} , inplace = True)
```

```python
data_final = pd.merge(data_final , data4 , how = 'outer')
```

```python
data1 = pd.read_csv("data1.csv" , encoding="ISO-8859-1")
```

```python
data1
```

```python
import codecs
def hash_rep(error) :
    return ("###" , error.start+1)
codecs.register_error('hash_rep' , hash_rep)
new_strs = []
for i in range(len(data1)): 
    new_strs.append(data1["Sentence"][i].encode('utf-8' , errors = 'hash_rep').decode('utf-8'))
```

```python
import re
for idx , s in enumerate(new_strs):
    if re.search(r"###" , s):
        print(idx , s)
```

```python
data1 = pd.DataFrame({"Sentiment" : data1["Sentiment"] , "Sentence" : new_strs})
```

```python
data1
```

```python
data_final = pd.merge(data_final , data1 , how = 'outer')
```

```python
data_final.isnull().sum()
```

```python
data_final.dropna(inplace = True)
data_final["Sentiment"] = data_final["Sentiment"].str.lower()
```
